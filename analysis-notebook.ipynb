```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from pandas.plotting import scatter_matrix
import yfinance as yf
import datetime
```

```python
start="1999-01-01"
end="2023-04-01"

sp500_symbol = "^GSPC"
ftse_symbol = "^FTSE"
eu50_symbol = "^STOXX50E"
nikkei_symbol = "^N225"
symbols = [sp500_symbol, ftse_symbol, eu50_symbol, nikkei_symbol]
```

```python
def get_grouped_historical_data(symbols):
    data = yf.download(symbols, start, end)
    return data

def get_single_historical_data(symbol):
    data = yf.download(symbol, start, end)
    return data
```

```python
# Get grouped historical data
grouped_historical_data = get_grouped_historical_data(symbols)

# Get S&P 500 data
sp500_data = get_single_historical_data("^GSPC")

# Get FTSE data
ftse_data = get_single_historical_data("^FTSE")

# Get EU50 data
eu50_data = get_single_historical_data("^STOXX50E")

# Get Nikkei data
nikkei_data = get_single_historical_data("^N225")
```

## Historical Stock Prices

```python
# Plotting the historical stock prices
grouped_historical_data["Adj Close"].plot(figsize=(12, 6))
plt.title('Historical Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(symbols)
plt.show()
```

## Moving Averages for S&P 500

```python
sp500_data['MA50'] = sp500_data['Open'].rolling(50).mean()
sp500_data['MA200'] = sp500_data['Open'].rolling(200).mean()
sp500_data['Open'].plot(figsize = (12,6))
sp500_data['MA50'].plot()
sp500_data['MA200'].plot()
plt.title('Moving Averages - S&P 500')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

## Linear regression for S&P 500

```python
# Extract the adjusted close prices and dates
adj_close = sp500_data['Adj Close']
dates = mdates.date2num(adj_close.index)

# Create a scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(x=dates, y=adj_close, alpha=0.5)

# Add a linear regression line
sns.regplot(x=dates, y=adj_close, scatter=False, color='red')

# Set the x-axis date formatting
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# Set the plot title and labels
plt.title('Historical Stock Prices of S & P 500 with Linear Regression')
plt.xlabel('Date')
plt.ylabel('Price')

# Show the plot
plt.show()
```
